# Bar Graph of Gender count

```python
# Bar Plot
gender_counts = df['Gender'].value_counts()
gender_counts.plot(kind='bar',cmap='Accent')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# Barplot of ratings of Foods on the basis of gender

```python
ratings = df.groupby('Gender')[['Burger', 'Juice', 'Bakeries', 'Cafes']].mean()

# Transpose the DataFrame for easier plotting
ratings = ratings.T

# Create a bar plot
ratings.plot(kind='bar', figsize=(10, 6)) # width,height
plt.title('Comparison of Ratings by Gender')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)  # Optional: Rotate the x-axis labels if needed

plt.show()
```

# Histogram of Distribution of Hotel Ratings

# Histogram of Average Ratings on Mall

```python
# Create a histogram for average ratings on resorts
plt.figure(figsize=(8, 6))
sns.histplot(df['Malls'], bins=20, kde=True,color='purple')
plt.title("Histogram of Average Ratings on Mall")
plt.xlabel("Ratings")
plt.show()
```

# Histplot of Ratings Comparison of Hotels, Restaurants, and Resorts

```python
# Extract the columns for hotels, restaurants, and resorts
hotels_ratings = df['Hotels']
restaurants_ratings = df['Restaurants']
resorts_ratings = df['Resort']

# Create a histogram for each rating
plt.figure(figsize=(10, 6))
plt.hist(hotels_ratings, bins=20, alpha=0.5, label='Hotels',color='red')
plt.hist(restaurants_ratings, bins=20, alpha=0.5, label='Restaurants',color='blue')
plt.hist(resorts_ratings, bins=20, alpha=0.5, label='Resort',color='green')

plt.title('Ratings Comparison: Hotels, Restaurants, and Resorts')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.legend()
plt.show()
```

# Piechart of Gender 

```python
# Pie Chart
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90) #autopct defines percentage
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

# Pie plot of Services Ratings

```python
# 'Services' is numeric
df['Services'] = pd.to_numeric(df['Services'], errors='coerce')  # 'coerce' handles non-numeric values

# Group the data by 'Gender' and calculate the sum of 'Services' ratings
services_by_gender = df.groupby('Gender')['Services'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(services_by_gender, labels=services_by_gender.index, autopct='%1.1f%%', startangle=90)
plt.title('Services Ratings by Gender')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.show()
```

# Lineplot

```python
#Lineplot
plt.figure(figsize=(10, 6))

# Filter data for males and females
male = df[df['Gender'] == 'Male']
female = df[df['Gender'] == 'Female']

# Line plot for Beauty ratings
sns.lineplot(x='ID', y='Beach', data=male, label='Male',color='b')
sns.lineplot(x='ID', y='Beach', data=female, label='Female',color='deeppink')

plt.title('Comparison of Ratings for Beach by Gender')
plt.xlabel('ID')
plt.ylabel('Ratings')
plt.legend(title='Gender')
plt.show()
```

# Scatter plot between average ratings on Swimming and gyms

# Scatter Plot of 'Park' and 'Gardens'


```python
# Create custom color palettes for 'Park' and 'Gardens' based on Gender
palette_park = {'Male': 'red', 'Female': 'blue'}
palette_gardens = {'Male': 'green', 'Female': 'purple'}

# Filter data for 'Park' and 'Gardens'
data_park = df[df['Park'] == 1]
data_gardens = df[df['Gardens'] == 1]

# Create scatter plots with custom colors for 'Park' and 'Gardens' by Gender
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Park', y='Gardens', data=data_park, hue='Gender', palette=palette_park)
sns.scatterplot(x='Park', y='Gardens', data=data_gardens, hue='Gender', palette=palette_gardens)

plt.title('Scatter Plot: Park vs. Gardens by Gender')
plt.xlabel('Park')
plt.ylabel('Gardens')
plt.legend(title='Gender')
plt.show()
```
